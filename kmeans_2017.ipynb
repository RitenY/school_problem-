```python
#PCAで次元削減した2017年の中学生のデータを取り込み
import pandas as pd
df1 = pd.read_csv('PCA_2017.csv')
df1.head()
```

```python
#クラスタリングに必要ないので都道府県列を削除
df2 = df1.drop(['Unnamed: 0', '2017'], axis = 1)
df2.head()
```

```python
#クラスを3つに指定してクラスタリングを実行
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3, random_state = 0)
model.fit(df2)
clusters = model.labels_
clusters
```

```python
#都道府県を戻してクラスタリングを確認
df2['cluster'] = clusters
df2['2017'] = df1['2017']
df2
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

#PCAでクラスタリングしたデータをプロット
plt.scatter(df2['PC1'], df2['PC2'], c = clusters) 
plt.title('PCA_plot')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()
clusters
```

```python
plt.scatter(df2['PC1'], df2['PC3'],c=clusters) 
plt.title('PCA_plot')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.colorbar()
plt.show()
```

```python
#クラスタごとのそれぞれ新規の特徴量の平均を出してグラフに表示
cluster_mean = df2.groupby('cluster').mean()
cluster_mean.plot(kind = 'bar', cmap = 'Accent')

#PC1 体力、規範意識、生活習慣、自己肯定感が高い
#PC2 不登校率が高い
#PC3 学力が低い
```

```python
#T-SNEで次元削減したデータでクラスタリング
#T-SNEで次元削減した2017年の中学生のデータを取り込み
df3 = pd.read_csv('tsne_2017.csv')

#クラスタリングに必要ないので都道府県列を削除
df4 = df3.drop(['Unnamed: 0', '2017'], axis = 1)
df4.head()
```

```python
#クラスを3つに指定してクラスタリングを実行
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3, random_state = 0)
model.fit(df4)
clusters = model.labels_
clusters
```

```python
#都道府県を戻してクラスタリングを確認
df4['cluster'] = clusters
df4['2017'] = df3['2017']
df4
```

```python
#T-SNEでクラスタリングしたデータをプロット
plt.scatter(df4['PC1'], df4['PC2'], c = clusters) 
plt.title('TSNE_plot')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
plt.show()
```

```python
#クラスタごとのそれぞれ新規の特徴量の平均を出してグラフに表示
cluster_mean = df4.groupby('cluster').mean()
cluster_mean.plot(kind = 'bar', cmap = 'Accent')
```
